# ATTENDANCE_RECORD

```python
import numpy as np
import pandas as pd
```

```python
#importing csv files
df1=pd.read_csv("r.csv")
df1
```

```python
import random
df1.dropna(inplace=True)   #Data frame is read and all rows with any NULl VALUES  are dropped.
df1.replace(' ---',round(random.uniform(75,100),2),inplace=True)  #replacing with randon uniform
df1.head()
```

# FILTERING OUT THE STUDENT WITH LESS THAN 75% 

```python
import numpy as np
filtering_out=[]                           #creating filtering out empty list
for i in range(0,df1.shape[0]):
    for j in range(4,32):
        if float(df1.iloc[i].values[j])<75:
            if i not in filtering_out:
                filtering_out.append(i)      #appending in list

#print(l)
#df2.head()
#df2.describe()
```

# CREATING THE DATA FRAME OF THE LESS THAN 75% ATTENDANCE

```python
#print(df1.columns[0])
Data_frame=[]                           #creating dataframe empty list 
make= pd.DataFrame()
for j in range(df1.shape[1]):
    for i in range(df1.shape[0]):
        if i in filtering_out:
            Data_frame.append(df1.iloc[i].values[j])
            make[df1.columns[j]]=Data_frame
            Data_frame.clear()
#make.head()
```

# FILTERING OUT THE STUDENT WITH MORE THAN 75% 

```python
filtering_out.clear()   
Data_frame.clear()
```

```python
flag=0
for i in range(0,df1.shape[0]):
    for j in range(4,28):
        if float(df1.iloc[i].values[j])<75:
            flag = 1
            continue
        if flag == 0 :
            if i not in filtering_out:
                filtering_out.append(i)
                flag = 0 
#print(df2.columns[0])
```

# CREATING THE DATA FRAME OF THE MORE THAN 75%  ATTENDANCE

```python
Data_frame=[]
attendance_OK= pd.DataFrame()
for j in range(df1.shape[1]):
    for i in range(df1.shape[0]):
        if i in filtering_out:
            Data_frame.append(df1.iloc[i].values[j])
            attendance_OK[df1.columns[j]]=Data_frame
            Data_frame.clear()
#attendance_OK.head()
```

```python
make.to_csv('make.csv') #detain list
attendance_OK.to_csv('attendance_OK.csv')   #more than 75% attendence list
```
